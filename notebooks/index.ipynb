````python
# Enhanced Ring Platform Testing Utilities with WebSocket Support
import requests
import json
import websocket
import threading
import time
from typing import Dict, List, Optional, Callable
import asyncio

class RingPlatformTester:
    """Comprehensive testing utilities for Ring Platform APIs and WebSocket features"""
    
    def __init__(self, base_url: str = "http://localhost:3000", api_key: str = None):
        self.base_url = base_url
        self.api_base = f"{base_url}/api"
        self.ws_url = f"ws://localhost:3000"  # WebSocket URL
        self.api_key = api_key
        self.session = requests.Session()
        self.ws_connection = None
        self.ws_messages = []
        
        # Set default headers
        if api_key:
            self.session.headers.update({"Authorization": f"Bearer {api_key}"})
        
        self.session.headers.update({
            "Content-Type": "application/json",
            "User-Agent": "Ring-Platform-Interactive-Docs/1.0"
        })
    
    def test_api_endpoint(self, endpoint: str, method: str = "GET", data: Dict = None, params: Dict = None):
        """Test any Ring Platform API endpoint with comprehensive error handling"""
        url = f"{self.api_base}{endpoint}"
        
        try:
            if method.upper() == "GET":
                response = requests.get(url, params=params, headers=self.session.headers)
            elif method.upper() == "POST":
                response = requests.post(url, json=data, headers=self.session.headers)
            elif method.upper() == "PUT":
                response = requests.put(url, json=data, headers=self.session.headers)
            elif method.upper() == "DELETE":
                response = requests.delete(url, headers=self.session.headers)
            else:
                raise ValueError(f"Unsupported HTTP method: {method}")
            
            return {
                "status_code": response.status_code,
                "success": response.status_code < 400,
                "data": response.json() if response.content else None,
                "headers": dict(response.headers),
                "url": url,
                "method": method.upper()
            }
        except Exception as e:
            return {
                "status_code": 0,
                "success": False,
                "error": str(e),
                "url": url,
                "method": method.upper()
            }
    
    def test_websocket_connection(self, auth_token: str = None):
        """Test WebSocket connection to Ring Platform real-time features"""
        def on_message(ws, message):
            self.ws_messages.append({
                "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "message": json.loads(message) if message.startswith('{') else message
            })
            print(f"üì° WebSocket Message: {message}")
        
        def on_error(ws, error):
            print(f"‚ùå WebSocket Error: {error}")
        
        def on_close(ws, close_status_code, close_msg):
            print("üîå WebSocket Connection Closed")
        
        def on_open(ws):
            print("‚úÖ WebSocket Connection Opened")
            # Send authentication if token provided
            if auth_token:
                ws.send(json.dumps({
                    "type": "auth",
                    "token": auth_token
                }))
        
        try:
            self.ws_connection = websocket.WebSocketApp(
                f"{self.ws_url}/api/socket",
                on_open=on_open,
                on_message=on_message,
                on_error=on_error,
                on_close=on_close
            )
            
            # Run WebSocket in a separate thread
            ws_thread = threading.Thread(target=self.ws_connection.run_forever)
            ws_thread.daemon = True
            ws_thread.start()
            
            return True
        except Exception as e:
            print(f"‚ùå Failed to connect to WebSocket: {e}")
            return False
    
    def send_websocket_message(self, message_type: str, data: Dict):
        """Send a message through WebSocket connection"""
        if self.ws_connection:
            payload = {
                "type": message_type,
                "data": data,
                "timestamp": datetime.now().isoformat()
            }
            self.ws_connection.send(json.dumps(payload))
            return True
        return False
    
    def test_ring_platform_health(self):
        """Comprehensive health check of Ring Platform services"""
        health_results = {}
        
        # Test main API
        health_results["main_api"] = self.test_api_endpoint("/")
        
        # Test authentication endpoints
        health_results["auth_status"] = self.test_api_endpoint("/auth/session")
        
        # Test major feature endpoints
        endpoints_to_test = [
            "/conversations",
            "/notifications", 
            "/news",
            "/entities",
            "/opportunities",
            "/wallet",
            "/admin/users"
        ]
        
        for endpoint in endpoints_to_test:
            service_name = endpoint.replace("/", "").replace("-", "_") or "root"
            health_results[f"{service_name}_api"] = self.test_api_endpoint(endpoint)
        
        return health_results
    
    def generate_test_report(self, results: Dict):
        """Generate a formatted test report"""
        print("üîç Ring Platform API Test Report")
        print("=" * 60)
        
        successful_tests = 0
        total_tests = len(results)
        
        for service, result in results.items():
            status = "‚úÖ PASS" if result.get("success", False) else "‚ùå FAIL"
            status_code = result.get("status_code", "N/A")
            
            print(f"{status} {service.upper():.<20} Status: {status_code}")
            
            if result.get("success", False):
                successful_tests += 1
            elif result.get("error"):
                print(f"    Error: {result['error']}")
        
        print("=" * 60)
        print(f"üìä Results: {successful_tests}/{total_tests} tests passed")
        print(f"üéØ Success Rate: {(successful_tests/total_tests)*100:.1f}%")
        
        return {
            "total_tests": total_tests,
            "successful_tests": successful_tests,
            "success_rate": (successful_tests/total_tests)*100
        }

# Initialize the tester
ring_tester = RingPlatformTester()

print("üîß Ring Platform Testing Utilities Initialized")
print("=" * 60)
print("üì° WebSocket Testing: Available")
print("üîç API Testing: Ready")
print("üìä Health Monitoring: Active")
print("üéØ Interactive Documentation: Online")
print("=" * 60)

# Quick platform health check
print("\nüè• Running Platform Health Check...")
health_results = ring_tester.test_ring_platform_health()
ring_tester.generate_test_report(health_results)
````
